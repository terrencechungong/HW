```python
def remove_punctuation(word: str):
    word = word.lower()
    return word.translate(str.maketrans('', '', string.punctuation))
```

```python
# init  hashmap of type word_freq[class] = {word: count}
word_freq = {} # {class:defaultdict(int)}
# init overall freq counter (vocabulary) to filter {word: count}
vocabulary = defaultdict(int) # {word:int}
# for topic in topics: LOGIC HERE
parent_folder = f"{current_path}/newsgroup_data/20_newsgroups/"  
seen_first = False
TOTAL_WORD_COUNT_FOR_CLASS = 'TOTAL_WORD_COUNT_FOR_CLASS/TOPIC'
for root, dirs, files in os.walk(parent_folder):
    if not seen_first:
        print("yoo")
        seen_first = True
        continue
    topic = root.split('/')[-1]
    print(topic)
    word_freq[topic] = defaultdict(int)

    for file in files:
        file_path = os.path.join(root, file)
        
        # Open and read each file
        with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
            content = f.read()
            # Split content into words
            words = content.split()
            
            # Iterate through each word in the file
            for word in words:
                # Remove punctuation from the word
                cleaned_word = remove_punctuation(word)
                word_freq[topic][cleaned_word] += 1
                word_freq[topic][TOTAL_WORD_COUNT_FOR_CLASS] += 1
                vocabulary[cleaned_word] += 1



#   topic_docs = all docs in topic directory
#   for doc in topic_docs:
#      for word in doc:
#          word_freq[topic][word] += 1
#          overall[word] += 1
#
# sort words (ascending) by most frequent and remove them from the dict. keep all else
# for word in words: (excluding top freqs)
#   for topic in topics:
#       word_freq[topic][word] += 1
#
# 
# classification
# prior x probability of observing each word in the class MAX OF THIS ACROSS ALL CLASSES
# if correct, add 1 to correct. accuracy = correct/20k
```

```python
top_200 = set()
vocab = sorted(vocabulary.items(), key=lambda x: x[1], reverse=True)
index = 1
topics = set()
for word, count in vocab:
    if index <= 200:
        top_200.add(word)
        index += 1
        continue
    for topic in word_freq.keys():
        word_freq[topic][word] += 1

for topic in word_freq.keys():
    topics.add(topic)
```

```python
from decimal import Decimal, getcontext
getcontext().prec = 90
import numpy as np



parent_folder = f"{current_path}/newsgroup_data/20_newsgroups/"  
seen_first = False
correct = 0
total = 0
# Iterate through each directory in the folder
for root, dirs, files in os.walk(parent_folder):
    if not seen_first:
        # print("yoo")
        seen_first = True
        continue
    real_topic = root.split('/')[-1]

    # print(real_topic, len(files))
    for file in files:
        inference_results = []
        # print(file, "FILE")
        for topic in topics:
            # print(topic, "TOPIC")
            prob_word = np.log(0.05)
            file_path = os.path.join(root, file)
            
            # Open and read each file
            with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                content = f.read()
                # Split content into words
                words = content.split()
                # print("WORDS", len(words))
                
                # Iterate through each word in the file
                for word in words:
                    # print(word, "WORDD")
                    # Remove punctuation from the word
                    cleaned_word = remove_punctuation(word)
                    if cleaned_word in top_200:
                        continue
                    # print(word_freq[topic][cleaned_word], cleaned_word, topic)
                    if word_freq[topic][cleaned_word] <= 0:
                        print("ZERO", topic, cleaned_word)
                    # print((word_freq[topic][cleaned_word] / word_freq[topic][TOTAL_WORD_COUNT_FOR_CLASS]), word_freq[topic][cleaned_word], word_freq[topic][TOTAL_WORD_COUNT_FOR_CLASS])
                    # print(np.log(word_freq[topic][cleaned_word] / word_freq[topic][TOTAL_WORD_COUNT_FOR_CLASS]))
                    prob_word += np.log(word_freq[topic][cleaned_word] / word_freq[topic][TOTAL_WORD_COUNT_FOR_CLASS])
                    # print(prob_word)
                inference_results.append([topic, prob_word])
        # print(real_topic, inference_results, file_path)
        highest_topic = (float('-inf'), "")
        for topic, prob_word in inference_results:
            # print(prob_word)
            if prob_word > highest_topic[0]:
                highest_topic = (prob_word, topic)
        if highest_topic[1] == real_topic:
            correct += 1
        total += 1
        print(correct, "CORRECT" , highest_topic, real_topic, correct/total)

print(f"Accuracy: {correct / 20000}")
# find what the max is and check the topic against
```
